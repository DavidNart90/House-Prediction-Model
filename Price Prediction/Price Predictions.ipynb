### Outliers

From the plot, we can see there are alot of outliers in the price column, let try and fix that. First let plot the distribution of the price colmn to know which type of outlier manipulation to use.

```python
df.boxplot(column="price") #boxplot to detect outliers
```

```python
sns.histplot(df["price"],kde=True)
```

From the distribution of the target variable, we will use the IQR method since it is right skewed and delete extreme bounderies.

```python
#Calculate the upper and lower bounderies of Price
quartile_1, quartile_3 = np.percentile(df.price, [25, 75])
iqr = quartile_3 - quartile_1
lower_bound = quartile_1 - (iqr * 1.5) #lower boundery
upper_bound = quartile_3 + (iqr * 1.5) # upper boundery
```

```python
# Delete the outliers
df = df[df.price<upper_bound]
df = df[df.price>lower_bound]
```

```python
df.shape
```

```python
# Now it seems a bit ok
sns.histplot(df["price"],kde=True)
```

#  Feature Selection and Engineering

```python
# Plotting the correlation matrix
plt.figure(figsize = (12,7))
sns.heatmap(round(df.corr(),3),annot=True, fmt="g",cmap="coolwarm")
```

- The correlation between price(target) and the other independent variable are very weak
- There are no importance between  the latitude, longitude and price so we drop those two features
- There are also no importance between id and host_id but we drop host_id and leave the id for our prediction file

```python
#drop unwanted columns
df.drop(["host_id","name","host_name","longitude","latitude","neighbourhood"],axis=1,inplace=True)
```

```python
df.head()
```

**One Hot-Encoding**

```python
#Treating Categorical Variables as Dummies
a = pd.get_dummies(df[["neighbourhood_group","room_type"]],drop_first=True)
#drop the categorical columns
df.drop(["neighbourhood_group","room_type"],axis=1,inplace=True)
```

```python
#concatenate dummies dataframe and df
df = pd.concat([df,a],axis=1)
```

```python
df.head()
```

## Model Training

```python
# Declaring the target variable(Y) and the independent variables(X) for training
X = df.drop(["id","price"],axis = 1)
y = df["price"]
```

```python
#Spitting data into training set and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) #30/70 split ratio
```

```python
#Base Model
from sklearn.linear_model import LinearRegression,  HuberRegressor
#Initialise the regression model
lm = LinearRegression()
# fit the model
lm.fit(X_train,y_train)
```

```python
#print coeffecients of the regression model
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df.head()
```

### Model Evaluation

```python
#import the evaluation metrics
from sklearn import metrics
# Make predictions
predictions = lm.predict(X_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print("R2 Score:", metrics.r2_score(y_test, predictions))
```

**Discussions**

- From the base model, 45% of the variations in price (y) is explained by the independent variables.
- The performance of the model is sorely based on the low correlations between the independent variable and the target variable.

```python
#Plot of the line of best fit
plt.figure(figsize=(10,7))
sns.regplot(y=y_test, x=predictions, color='blue')
plt.title('Actual vs Predicted', fontsize=15)
plt.xlabel('Predictions',fontsize=10)
plt.ylabel('Actual', fontsize = 10)
```

-        The graph shows the best fit model, which plots the actuals vs the predicted variables, which suggests that there is a moderately strong linear relationship (r-squared = 45.5%) between the price and the other independent variables.

```python
#Plot the residuals
plt.figure(figsize=(10,7))
sns.histplot((y_test-predictions),bins=50,kde=True)
plt.title('Residuals', fontsize=15)
```

- From the residual plot, we can see the difference between actual and the predicted y's is normally distributed. Which implies that the model is quite okay for predictions.

### Prediction of Prices

```python
# Predict the prices
df['Price Prediction'] = lm.predict(df[X_test.columns])[:]
```
